```python
tmp_df = pd.merge(df, user_pref_df, on=['match_year', 'match_round','home_team','away_team'])
tmp_df.head()
```

```python
final_df = tmp_df.drop(['match_date','home_team','away_team','home_goals','away_goals'], axis=1)
final_df = final_df.sort_values(by = ['match_year', 'match_month','match_day','match_hour','match_round']).reset_index(drop=True)
final_df.head()
```

```python
print(f"# pareggi : {list(final_df.label).count(0)} ({list(final_df.label).count(0)*100/final_df.shape[0]:.2f})")
print(f"# vittorie casa : {list(final_df.label).count(1)} ({list(final_df.label).count(1)*100/final_df.shape[0]:.2f})")
print(f"# vittorie fuori casa : {list(final_df.label).count(2)} ({list(final_df.label).count(2)*100/final_df.shape[0]:.2f})")
```

```python
final_df = final_df[['10bet_1','10bet_2','10bet_X','10bet_opening_1','10bet_opening_2','10bet_opening_X','at_win_pred_transfer','draw_pred_transfer','ht_win_pred_transfer','match_day','match_month','match_year','label']]
```

```python
import lightgbm as lgb

from sklearn.model_selection import train_test_split

X = final_df.drop(['label'],axis=1)
y = final_df.label

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,shuffle=False)

model = lgb.LGBMClassifier(learning_rate=0.1,
                          num_leaves=5, n_estimators=100, reg_lambda=0.05,
                           reg_alpha=0.0
                          )
model.fit(X_train, y_train)

# mio modello
predicted_y = model.predict(X_test)
correct = 0
for ind,true,predicted in zip(y_test.index, y_test, predicted_y):
    if true == predicted:
        correct+=1
correct*100 / y_test.shape[0]
```

```python
# baseline che usa solo le quote
result = []
for p1,pX,p2 in zip(X_test['10bet_1'], X_test['10bet_X'], X_test['10bet_2']):
    if min([p1,pX,p2]) == p1: result.append(1)
    elif min([p1,pX,p2]) == pX: result.append(0)
    else: result.append(2)

correct = 0
for ind,true,predicted in zip(y_test.index, y_test, result):
    if true == predicted:
        correct+=1
correct*100 / y_test.shape[0]
```

```python
# baseline che usa solo le predictions
result = []
for p1,pX,p2 in zip(X_test['ht_win_pred_transfer'], X_test['draw_pred_transfer'], X_test['at_win_pred_transfer']):
    if max([p1,pX,p2]) == p1: result.append(1)
    elif max([p1,pX,p2]) == pX: result.append(0)
    else: result.append(2)

correct = 0
for ind,true,predicted in zip(y_test.index, y_test, result):
    if true == predicted:
        correct+=1
correct*100 / y_test.shape[0]
```
