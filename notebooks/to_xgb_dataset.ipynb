```python
import pandas as pd
df = pd.read_csv('../dataset/serie_a/transformed/transformed_merged_serie_a.csv')
df.head()
```

```python
label = []
for hg,ag in zip(df.home_goals, df.away_goals):
    if hg == ag:
        label.append(0)
    elif hg > ag:
        label.append(1)
    else:
        label.append(2)
df['label'] = label
df.head(20)
```

```python
for i in range(df.shape[0]):
     df.at[i, 'match_hour'] = int(df.at[i, 'match_hour'].split(':')[0])
```

```python
df
```

```python
match_month, match_day, match_year = [], [], []
for date in df.match_date:
    match_day.append(int(date.split('.')[0]))
    match_month.append(int(date.split('.')[1]))
    match_year.append(int(date.split('.')[2]))
```

```python
df['match_day'] = match_day
df['match_month'] = match_month
df['match_year'] = match_year
df.match_hour = df.match_hour.astype(int)
```

```python
final_df = df.drop(['match_date','home_team','away_team','home_goals','away_goals'], axis=1)
final_df = final_df.sort_values(by = ['match_year', 'match_month','match_day','match_hour','match_round']).reset_index(drop=True)
final_df.head()
```

```python
print(f"# pareggi : {list(final_df.label).count(0)} ({list(final_df.label).count(0)*100/final_df.shape[0]:.2f})")
print(f"# vittorie casa : {list(final_df.label).count(1)} ({list(final_df.label).count(1)*100/final_df.shape[0]:.2f})")
print(f"# vittorie fuori casa : {list(final_df.label).count(2)} ({list(final_df.label).count(2)*100/final_df.shape[0]:.2f})")
```

```python
import lightgbm as lgb
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_df.drop(['label'],axis=1)
y = final_df.label
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,shuffle=False)
```

```python
model = lgb.LGBMClassifier(learning_rate=0.01,
                          num_leaves=3
                          )
model.fit(X_train, y_train)
```

```python
predicted_y = model.predict(X_test)
correct = 0
for ind,true,predicted in zip(y_test.index, y_test, predicted_y):
    if true == predicted:
        correct+=1
correct*100 / y_test.shape[0]
```
