```python
import pandas as pd
odds_df = pd.read_csv('../dataset/premier/lightgbm/lightgbm_base_odds_premier.csv')
odds_df = odds_df.sort_values(by=['match_year','match_month','match_day','match_hour','match_round','home_team','away_team']).reset_index(drop=True)
odds_df.head()
```

```python
bet365 = odds_df[['bet365_1','bet365_X','bet365_2','label','match_day','match_month','match_year','match_hour','match_round']]
bet365.head()
```

```python
# baseline che usa solo le quote
subset_bet365 = bet365[(bet365.match_round<34) & (bet365.match_year>2016)].reset_index(drop=True)
result = []
for p1,pX,p2 in zip(subset_bet365['bet365_1'], subset_bet365['bet365_X'], subset_bet365['bet365_2']):
    if min([p1,pX,p2]) == p1: result.append(1)
    elif min([p1,pX,p2]) == pX: result.append(0)
    else: result.append(2)

correct = 0
for ind,true,predicted in zip(subset_bet365.index, subset_bet365.label, result):
    if true == predicted:
        correct+=1
correct*100 / subset_bet365.shape[0]
```
